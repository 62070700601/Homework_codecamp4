# Quiz1

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Company = pd.Series(['A','B','C','D'])
Year = pd.Series(['2014','2015','2016','2017'])
income = np.random.uniform(low=50*10**6, high=500*10**6, size=(4,4))
df = pd.DataFrame(income,Company,Year)
df
```

# Quiz2

```python
df['Avg']=(df['2014']+df['2015']+df['2016']+df['2017'])/4
df['Avg']
fig = plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
sns.barplot(x=df.index ,y='Avg',data=df)

plt.subplot(1,2,2)
sns.barplot(x='Avg',y=df.index,data=df)
```

# Quiz3

```python
fig = plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
sns.scatterplot(x= df.index,y = df['2017'],data=df)
plt.subplot(1,2,2)
sns.scatterplot(x = df['2017'],y= df.index,data=df)
```

# Quiz13

```python
# สร้าง HeatMap โดย crosstab กำหนดให้ index เป็น Pclass และ column เป็นเพศ 
df13 = pd.read_csv('train.csv')
dfnew13 = pd.crosstab(df13['Pclass'],df13['Sex'])
dfnew13
sns.heatmap(dfnew13,cmap='twilight')
```

# Quiz14 

```python
# สร้าง HeatMap โดย pivot table กำหนดให้ index เป็น Pclass และ column เป็นเพศ และ values เป็น ราคา
df14 = pd.read_csv('train.csv')
newdf14= df14.pivot_table(index='Pclass',columns='Sex',values='Fare')
sns.heatmap(newdf14,cmap='coolwarm')
```

 # Quiz15

```python
# สร้าง HeatMap โดย pivot table กำหนดให้ index เป็น Pclass และ column เป็นเพศ และ values เป็น ราคา และกำหนดเส้นสีขาวขนาด 2 เป็นเส้นแบ่ง
df15 = pd.read_csv('train.csv')
dfnew15 = df15.pivot_table(index='Pclass',columns = 'Sex',values = 'Fare')
sns.heatmap(dfnew15,linecolor='white',linewidths=0.5,cmap='PiYG')
```
