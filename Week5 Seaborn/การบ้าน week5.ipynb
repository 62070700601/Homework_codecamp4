# Quiz9 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('train.csv')
AllValue = df['Pclass'].count()
Pclass1 = df[df['Pclass']==1]
Pclass1_value = Pclass1['Pclass'].count()
print('จำนวน Pclass1 =',Pclass1_value,'ตัว')
Pclass2 = df[df['Pclass']==2]
Pclass2_value = Pclass2['Pclass'].count()
print('จำนวน Pclass2 =',Pclass2_value,'ตัว')
Pclass3 = df[df['Pclass']==3]
Pclass3_value = Pclass3['Pclass'].count()
print('จำนวน Pclass3 =',Pclass3_value,'ตัว')
Percentage1 = (Pclass1_value*100)/AllValue
Percentage2 = (Pclass2_value*100)/AllValue
Percentage3 = (Pclass3_value*100)/AllValue
AllPercentage = [Percentage1,Percentage2,Percentage3]
plt.figure(figsize =  [5,5])
label = ['Pclass1 %','Plclass2 %','Pclass3 %']
plt.pie(AllPercentage,labels = label,autopct='%1.1f%%')
plt.title('Percentage of Pclass')
plt.show()
```

# Quiz10

```python
## ให้หาอัตราส่วนของจุดหมายปลายทาง (Embark) และพล็อต Pie chat พร้อมทั้งให้ Queenstown เป็น explode
df10 = pd.read_csv('train.csv')
Allvalue = df10['Embarked'].count()
Cherbourg = df10[df10['Embarked'] == 'C']
CherbourgCount = Cherbourg['Embarked'].count()
Queenstown = df10[df10['Embarked'] == 'Q']
QueenstownCount = Queenstown['Embarked'].count()
Southampton = df10[df10['Embarked'] == 'S']
SouthamptonCount = Southampton['Embarked'].count()
Cratio = (CherbourgCount*100)/Allvalue
Qratio = (QueenstownCount*100)/Allvalue
Sratio = (SouthamptonCount*100)/Allvalue
plt.figure(figsize =  [6,6])
expode = [0,0,0.05]
plt.pie([Cratio,Qratio,Sratio],labels = ['Cherbourg','Queenstown','Southampton'],explode = expode,autopct='%1.1f%%')
```
