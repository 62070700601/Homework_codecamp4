```python
# ข้อที่ 12

from sklearn.model_selection import train_test_split

x = df.drop('class',axis=1)
y = df['class']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
# ข้อที่ 14
# ใช้ Normallization เพราะข้อมูลไม่ได้เป็น normal distribution
```

```python
# ข้อที่ 15
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
arr_minmax = min_max_scaler.fit_transform(df.drop('class',axis=1))
arr_minmax
```

```python
df_minmax = pd.DataFrame(arr_minmax,columns=df.columns[1:])
df_minmax
```

```python
x_train,x_test,y_train,y_test = train_test_split(df_minmax,df['class'],test_size=0.2,random_state=30)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train,y_train)
predicted = knn.predict(x_test)
predicted
```

```python
# ข้อที่ 16

from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predicted)
```

```python
print('Accuracy = ', accuracy_score(y_test,predicted))
print('F1 score = ', f1_score(y_test,predicted,average='micro'))
print('Precision score = ', precision_score(y_test,predicted,average='micro'))
print('Recall score = ', recall_score(y_test,predicted,average='micro'))
```

```python
# ข้อที่ 17
k_margin = []

for i in range(1,30):
    knn = KNeighborsClassifier(n_neighbors= i)
    knn.fit(x_train,y_train)
    predicted_i = knn.predict(x_test)
    k_margin.append(accuracy_score(y_test,predicted_i))

plt.figure(figsize=(10,8))
plt.plot(range(1,30),k_margin,color='black',linestyle = 'dashed',marker='o' ,markerfacecolor = 'blue',markersize=7)
plt.xlabel('K')
plt.ylabel('Accuracy')
    
```

```python
knn = KNeighborsClassifier(n_neighbors=28)
knn.fit(x_train,y_train)
predicted = knn.predict(x_test)
confusion_matrix(y_test,predicted)
```

```python
print('Accuracy = ',accuracy_score(y_test,predicted))
```
