# ข้อที่ 4.20

```python
#20. done แสดงเป็นกลุ่มช่วงอายุ 0-10, 20-30 …. (เพิ่มทีละ10) และหาว่าแต่ละกลุ่มมีอัตราการรอดชีวิตเท่าไหร่ 
df_point20=pd.read_csv('train.csv')
age_range=df_point20['Age'].apply(lambda x:((x//10)*10))

age_range2=df_point20['Age'].apply(lambda x:((x//10)*10)+9    )
age_range2
df_point20['Age1']=pd.Series(age_range)
df_point20['Age2']=pd.Series(age_range2)

df_point20['Age1']=df_point20['Age1'].astype(str)
df_point20['Age2']=df_point20['Age2'].astype(str)

df_point20['AgeGroup'] = df_point20['Age1']+"-"+df_point20['Age2']
df_point20_1 = df_point20[df_point20['Survived'] == 1]

df_point20_2['CountSurvived'] = pd.DataFrame(df_point20_1.groupby(['AgeGroup'])['Survived'].count())

df_point20_2.sum()['CountSurvived']
df_point20_2['Survived Ratio']=df_point20_2['CountSurvived']*100/df_point20_2.sum()['CountSurvived']
df_point20_2
```
